```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar las corridas 'mbo' para una mejor visualización de los RMSE más pequeños
corridas_filtradas = corridas[~corridas['Corrida'].str.contains('(mbo)', regex=False)].copy()

# Convertir RMSE a tipo numérico, asegurándose de que no haya errores si ya es numérico
corridas_filtradas['RMSE'] = pd.to_numeric(corridas_filtradas['RMSE'])

# Ordenar por RMSE para una mejor visualización
corridas_filtradas = corridas_filtradas.sort_values(by='RMSE', ascending=True)

plt.figure(figsize=(12, 7))
sns.barplot(x='Corrida', y='RMSE', data=corridas_filtradas, palette='viridis')
plt.title('Comparación de RMSE por Corrida (excluyendo mbo)')
plt.xlabel('Corrida')
plt.ylabel('RMSE')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

El gráfico muestra los valores de RMSE para cada corrida, excluyendo las que tienen 'mbo' en su nombre para una mejor escala visual. Las barras están ordenadas de menor a mayor RMSE para facilitar la identificación de las corridas con mejor rendimiento.

## Extraer Parámetros 'learning_rate' y 'num_leaves'

### Subtask:
Utilizar expresiones regulares o manipulación de cadenas para extraer los valores numéricos de 'learning_rate' y 'num_leaves' de la columna 'Configuracion' del DataFrame `top_3_corridas`. Almacenar estos valores en nuevas columnas.


**Razonamiento: Para extraer los valores 'learning_rate' y 'num_leaves' de la columna 'Configuracion', utilizaré expresiones regulares para encontrar los patrones numéricos asociados con estas palabras clave y almacenarlos en nuevas columnas respectivas en el DataFrame top_3_corridas.




```python
import re

# Extraer learning_rate
top_3_corridas['learning_rate'] = top_3_corridas['Configuracion'].apply(lambda x: float(re.search(r'learning_rate=(\d+\.?\d*)', x).group(1)) if re.search(r'learning_rate=(\d+\.?\d*)', x) else None)

# Extraer num_leaves
top_3_corridas['num_leaves'] = top_3_corridas['Configuracion'].apply(lambda x: int(re.search(r'num_leaves=(\d+)', x).group(1)) if re.search(r'num_leaves=(\d+)', x) else None)

# Mostrar el DataFrame actualizado
display(top_3_corridas)
```

Razonamiento: El código anterior generó una advertencia SettingWithCopyWarning porque top_3_corridas era una sección de otro DataFrame, y la asignación directa a nuevas columnas podría generar un comportamiento impredecible. Para indicar explícitamente la intención de modificar el DataFrame y suprimir esta advertencia, usaré .loc para la asignación de columnas.


```python
import re

# Extraer learning_rate usando .loc para evitar SettingWithCopyWarning
top_3_corridas.loc[:, 'learning_rate'] = top_3_corridas['Configuracion'].apply(lambda x: float(re.search(r'learning_rate=(\d+\.?\d*)', x).group(1)) if re.search(r'learning_rate=(\d+\.?\d*)', x) else None)

# Extraer num_leaves usando .loc para evitar SettingWithCopyWarning
top_3_corridas.loc[:, 'num_leaves'] = top_3_corridas['Configuracion'].apply(lambda x: int(re.search(r'num_leaves=(\d+)', x).group(1)) if re.search(r'num_leaves=(\d+)', x) else None)

# Mostrar el DataFrame actualizado
display(top_3_corridas)
```

## Visualizar y Comparar Parámetros

### Subtask:
Generar un gráfico o una tabla comparativa para visualizar las relaciones entre 'learning_rate', 'num_leaves' y el 'RMSE' para las top 3 corridas. Asegúrate de incluir leyendas claras.


Razonamiento: Para visualizar las relaciones entre la tasa de aprendizaje, el número de hojas y el RMSE de las tres primeras ejecuciones, crearé un diagrama de dispersión. Este tipo de diagrama permite mostrar eficazmente tres variables numéricas al asignar dos a los ejes y la tercera al color de los puntos, lo que facilita la comparación según se solicite.
